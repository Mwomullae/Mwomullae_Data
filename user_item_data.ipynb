```python
import pandas as pd
```

```python
# windows에서 엑셀을 cp949방식으로 저장을 하기 때문에 encoding="utf-8"이 아니라 encoding="cp949" 이용해야함
data_set = pd.read_csv("/content/drive/MyDrive/dataset/restaurant_information_2.csv", encoding="cp949", usecols=[0, 1, 2, 3, 4, 5, 6, 7])
data_set.head(5)
```

---
# 02. List 생성: 원소 Random 추출 
---

---
* 특정 column 데이터를 리스트로 변환
---

```python
# restaurant_name(음식점 이름) data -> list
restaurant_list = []
restaurant_tmp = data_set["restaurant_name"].to_list()
print(len(restaurant_tmp))

# restaurant_name에 식별자 추가
for i in range(193):
    i = f"{i}_" + restaurant_tmp[i] 
    restaurant_list.append(i)

print(restaurant_list)
```

---
* 평점 데이터 생성 후 리스트로 변환
---

```python
# rating data -> list : 0, 1.0 ~ 5.0
rating_list = []

rating_list.append(0.0)
for i in range(1, 5):
        for j in range(0, 10):
            num = f"{i}.{j}"
            num = float(num)
            rating_list.append(num)

rating_list.append(5.0)
print(rating_list)
```

---
# 03. 랜덤 추출: 음식점 이름, 평점 
---


---
* user 400명
* user 1명당 100개의 음식점 이름 & 평점
---

```python
import random

rating_data = []
restaurant_data = []

user_id_data = []
user_rating_data = []
user_restaurant_data = []

for user in range(400): 
    for i in range(100): 
        user_id_data.append(user+1)
        
        rating = random.sample(rating_list, 1)
        restaurant = random.sample(restaurant_list, 1)

        for j in rating:
            user_rating_data.append(j)

        for j in restaurant:
            user_restaurant_data.append(j)

    rating_data  = []
    restaurant_data = []

print(user_id_data)
print(user_rating_data)
print(user_restaurant_data)
```

```python
print(len(user_id_data))
print(len(user_rating_data))
print(len(user_restaurant_data))
```

---
# 04. user_dataframe 생성 
---

```python
user_list = [ 
                ['user_id',user_id_data],
                ['rating',user_rating_data],
                ['restaurant',user_restaurant_data] 
              ]
user_data = pd.DataFrame.from_dict(dict(user_list))
user_data['restaurant_id'] = user_data.restaurant.str.split('_').str[0]
print(user_data)
```

```python
user_restaurant_data = []
restaurant_tmp = user_data["restaurant"].to_list()
print("수정 전:", restaurant_tmp)

for restaurant in restaurant_tmp:
    restaurant = restaurant[restaurant.find(" ")+1:]
    user_restaurant_data.append(restaurant)
print("수정 후:", user_restaurant_data)
```

```python
user_data = user_data.drop(['restaurant'],axis=1)
user_data
```

# 05. item_dataframe 생성 
---

```python
# restaurant_name(음식점 이름) data -> list
restaurant_list = []
restaurant_tmp = data_set["restaurant_name"].to_list()
print(len(restaurant_tmp))

# restaurant_name에 식별자 추가
for i in range(193):
    i = f"{i}_" + restaurant_tmp[i] 
    restaurant_list.append(i)

print(restaurant_list)
```

```python
item_list = [ 
                ['restaurant',restaurant_list] 
              ]
item_data = pd.DataFrame.from_dict(dict(item_list))
item_data['restaurant_id'] = item_data.restaurant.str.split('_').str[0]
```

```python
item_data
```

```python
item_restaurant_data = []
print("수정 전:", restaurant_list)

for restaurant in restaurant_list:
    restaurant = restaurant[restaurant.find(" ")+1:]
    item_restaurant_data.append(restaurant)
print("수정 후:", item_restaurant_data)
```

```python
item_data = item_data.drop(['restaurant'],axis=1)
item_data["restaurant"] = item_restaurant_data
```

```python
item_data
```

# 06. user_dataframe & item_dataframe 확인 및 merge 
---

```python
user_data
```

```python
item_data
```

```python
# merge: user_data & item_data
# merge: 공통 column[restaurant_id] 
user_item_data = pd.merge(user_data, item_data, on="restaurant_id")
user_item_data
```

```python
# user_id와 restaurant_id가 중복일 경우 
user_item_data = user_item_data.drop_duplicates(subset=['user_id','restaurant_id'])
user_item_data
```

```python
user_item_data = user_item_data[["user_id", "restaurant_id", "rating", "restaurant"]]
user_item_data
```
