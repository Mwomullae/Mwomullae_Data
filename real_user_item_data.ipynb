```python
import pandas as pd

dataset = pd.read_csv("/content/drive/MyDrive/dataset/user_item_real_data.csv", encoding="cp949")
dataset.head() 
```

---
# 02. Data Setting
---

```python
dataset = dataset.drop("타임스탬프", axis=1)
dataset.head()
```

```python
dataset.info()
```

```python
datset_col = dataset.columns.to_list()
print(datset_col)
```

```python
row_len = len(dataset["부산삼겹살 하단본점"])
```

```python
index = 0
count = 1
column_list = dataset.columns.to_list()
column_len = len(column_list) 



for i in range(column_len):
    if (count == 1):
        user_data = pd.DataFrame({f"{count}_rating":dataset[column_list[index]].values})
        length = len(user_data[f'{count}_rating'])
        index = index + 1
    else:
        user_data[f"{count}_rating"] = dataset[column_list[index]].values
        index = index + 1

    for _ in range(length):
        user_data[f"{count}_restaurant_id"] = count
    count = count + 1
user_data
```

```python
count = 1
user_id = []

for i in range(column_len):
    for i in range(row_len):
        user_id.append(count)
        count = count + 1
    count = 1
```

```python
# user_matrix 생성
count = 1
ration_data = []
restaurant_id_data = []

for i in range(35):
    i = user_data[f"{count}_rating"].to_list()
    ration_data.extend(i)

    i = user_data[f"{count}_restaurant_id"].to_list()
    restaurant_id_data.extend(i)
    count = count + 1

user_data = pd.DataFrame({"user_id": user_id, "restaurant_id":restaurant_id_data, "rating": ration_data})
user_data
```

```python
# item_matrix 생성
count = 1
restaurant = []
restaurant_id = []

for i in datset_col:
    restaurant.append(i)
    restaurant_id.append(count)
    count = count + 1

item_data = pd.DataFrame({"restaurant_id":restaurant_id, "restaurant": restaurant})
item_data
```

---
# 03. Data Merge
---

```python
# merge: user_data & item_data
# merge: 공통 column[restaurant_id] 
user_item_data = pd.merge(user_data, item_data, on="restaurant_id")
user_item_data
```
